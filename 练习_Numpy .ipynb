### 三元函数向量化形式

```python
xarr = np.array([1.1,1.2,1.3,1.4,1.5])  # x
yarr = np.array([2.1,2.2,2.3,2.4,2.5])  # y
cond = np.array([True,False,True,True,False])  # c
```

三元函数判断
注意: ①数组加[] ②for x,y,z in zip(xarr,yarr,cond)

```python
result = [(x if c else y)
          for x,y,c in zip(xarr,yarr,cond)]
result
```

三元表达式的向量化版本

```python
result = np.where(cond,xarr,yarr)
result
```

### 测试_实例:随机漫步
数组添加>: 注意是walk.append(position),   
      而不能walk = walk.append(position)  
**产生1000个随机整数**  
nsteps = 1000  
draws = np.random.randint(0,2,size=nsteps)   
steps = np.where(draws > 0, 1, -1)

```python
import random
position = 0
walk = [position]
steps = 1000

for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)  
```

```python
import matplotlib.pyplot as plt
plt.plot(walk[:100])
```

```python
"""
使用numpy模块
"""
nsteps = 1000
draws = np.random.randint(0,2,size=nsteps)  
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
print(walk.max(), walk.min())
```

```python
# Returns the indices of the maximum values along an axis.
(np.abs(walk) >= 10).argmax() 
```

```python
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0,2,size=(nwalks,nsteps))
steps = np.where(draws>0, 1 ,-1)
walks = steps.cumsum(1)  # 注意axis=1
walks
```

```python
print(walks.min(),walks.max())
```
